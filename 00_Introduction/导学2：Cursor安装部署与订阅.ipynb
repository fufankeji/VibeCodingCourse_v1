## 1. Cursor 的定位与核心价值

&emsp;&emsp;在正式进入安装之前，我们必须先解决一个关键问题：Cursor 到底是什么？它为什么在工程实践中会被认为是"AI IDE"，而不仅仅是"VS Code + AI 功能"？这一章的作用，就是帮助我们建立 Cursor 的宏观定位，并理解它相对于传统编辑器的本质差异。只有真正理解了这个差异，后续的安装、订阅和配置才有明确的目标感——我们不是在"装一个软件"，而是在搭建一个全新的 AI 协作研发环境。

&emsp;&emsp;本章将从两个维度展开：首先明确 Cursor 的官方定义与设计哲学，然后通过与传统编辑器的对比，把 AI 集成层级的差异讲透。

### 1.1 Cursor 的官方定义：AI-native IDE

&emsp;&emsp;Cursor 在官网的定位非常直接：它不是在编辑器里加一个聊天窗口，而是 **Built to make you extraordinarily productive**——"为 AI 编程而生的开发环境"。这句话的关键在于"原生"二字：AI 不是 Cursor 的附加功能，而是 Cursor 的核心能力。

&emsp;&emsp;这意味着 Cursor 的设计出发点不是"辅助写代码"，而是"AI 参与开发流程"。具体来说，Cursor 的 AI 能力覆盖了以下几个维度：

- **代码理解**：不是单文件级别的语法分析，而是全项目级的语义理解
- **上下文检索**：通过 `Maximum Context Window` 让 AI 看到更完整的项目结构
- **多文件修改**：支持 Agent 自动执行跨文件的重构与生成任务
- **任务自动执行**：通过 `Agent Mode` 让 AI 成为真正的"执行者"而不仅仅是"建议者"

&emsp;&emsp;<font color=red>这是 Cursor 与所有"编辑器 + AI 插件"方案的本质区别：AI 不是在旁边给建议，而是直接参与到研发流程的执行链路中。</font>

### 1.2 Cursor 相对传统编辑器的独特之处

&emsp;&emsp;当我们对比 `VS Code + Copilot` 与 Cursor 时，可以发现差异集中在 AI 的集成层级上：传统编辑器是外挂式增强，而 Cursor 是 IDE 内核级 AI 工作流。为了把这个差异讲清楚，我们用一张对比表来呈现：

|----------|-------------------|--------|

&emsp;&emsp;举个具体的例子：如果我们需要在一个 FastAPI 项目中新增一个 CRUD 模块，传统 Copilot 只能在你打开的单个文件里给补全建议；而 Cursor 的 Agent 可以自动识别项目结构，创建 router、schema、service 等多个文件，并保持它们之间的引用关系正确。

### 2.3 macOS 安装步骤

&emsp;&emsp;macOS 环境下 Cursor 提供 ARM64（Apple Silicon）与 x64（Intel）版本，下载时必须匹配芯片架构。如果你不确定自己的 Mac 是什么芯片，可以点击左上角苹果图标 → "关于本机"查看。

### 2.4 Linux 安装（Deb/RPM/AppImage）

&emsp;&emsp;在 Linux 环境下 Cursor 提供 `.deb`、`.rpm` 与 `AppImage` 三种格式，分别适配不同的发行版：

- **`.deb`**：适用于 Debian/Ubuntu 系列
- **`.rpm`**：适用于 Fedora/CentOS/RHEL 系列
- **`AppImage`**：通用格式，无需安装，直接运行

&emsp;&emsp;下面我们以 Ubuntu 为例演示 `.deb` 包的安装流程：

````python
# Debian/Ubuntu 系统安装 Cursor
# 假设已下载 cursor.deb 到当前目录

# 步骤 1：使用 dpkg 安装 deb 包
!sudo dpkg -i cursor.deb

# 步骤 2：修复可能缺失的依赖
!sudo apt-get install -f -y
````

&emsp;&emsp;上面这段命令做了两件事：首先通过 `dpkg` 直接安装 deb 包，然后通过 `apt-get install -f` 自动修复可能缺失的依赖库。执行后如果依赖完整，Cursor 将出现在应用列表中，启动即可。

&emsp;&emsp;**成功判据**：在终端输入 `cursor` 或从应用菜单启动，能正常打开编辑器界面。
