# <center>《Vibe Coding AI全栈开发实战》开发环境配置前置准备 </center>

&emsp;&emsp;本期教程我们聚焦的是 **Vibe Coding 环境配置的第一步——网络环境准备**。在正式安装 Cursor、Git 和配置 GitHub 之前，我们需要先确保网络环境能够支撑后续的所有操作。这一步看似简单，却是很多人在环境配置过程中遇到问题的根源。

&emsp;&emsp;为什么要把网络环境单独拿出来讲？因为 Vibe Coding 的核心工具链——Cursor、GitHub、AI 模型服务——都依赖稳定的国际网络连接。<font color=red>如果网络环境不通畅，后续的每一步都可能卡壳：Git clone 超时、Cursor 登录失败、Claude 模型无法调用……</font>所以，我们先把这个「地基」打好。

&emsp;&emsp;接下来的内容会按这样的主线展开：首先理解为什么 Vibe Coding 环境需要 VPN，然后检测你现有的网络环境是否满足要求，最后为没有 VPN 的同学推荐一个开发者友好的解决方案。完成这一课后，大家就可以带着「网络通畅」的信心进入 Cursor 和 Git 的安装配置了。

## 1. 为什么 Vibe Coding 环境需要 VPN？

&emsp;&emsp;在开始配置任何开发工具之前，我们需要先正视一个现实问题：**国内网络环境对 Vibe Coding 工作流的支持是不完整的**。这不是危言耸听，而是实际开发中每天都会遇到的痛点。让我们逐一分析这些场景。

### 1.1 GitHub 连接不稳定

&emsp;&emsp;GitHub 是全球最大的代码托管平台，也是 Vibe Coding 工作流中不可或缺的一环。无论是克隆开源项目、提交代码、还是与 Cursor Agent 协作进行版本管理，都需要与 GitHub 进行通信。

&emsp;&emsp;然而，国内直连 GitHub 的体验可以用「随缘」来形容：

- `git clone` 速度极慢，几 KB/s 的下载速度是常态
- 连接频繁中断，大型仓库根本无法完整克隆
- `git push` 经常超时失败，需要反复重试
- GitHub 网页有时无法正常加载

&emsp;&emsp;<font color=red>这种不稳定性会严重影响开发效率，甚至让某些操作完全无法进行。</font>

### 1.2 Cursor Pro 的 AI 模型需要 VPN

&emsp;&emsp;Cursor 作为 AI-native IDE，其核心竞争力在于内置的 AI 编程能力。订阅 Cursor Pro 后，你可以使用 Claude Opus4.5、GPT-5.2 等顶级 AI 模型进行代码生成、重构和问答。

&emsp;&emsp;但这里有一个关键点：<font color=red>这些 AI 模型的服务器都部署在海外，国内网络无法直接访问。</font>具体表现为：

- 按下 `Ctrl+I` 打开 Agent 面板后，请求长时间无响应
- Tab 补全功能时灵时不灵
- 模型切换为 Claude 后完全无法使用
- 出现 "Network Error" 或 "Request Timeout" 等错误

&emsp;&emsp;虽然 Cursor 支持通过自定义 API Key 接入国内模型（如 DeepSeek、Qwen），但如果你想使用 Cursor Pro 订阅自带的 Claude 模型额度，VPN 是必需的。

### 1.3 其他开发工具和资源

&emsp;&emsp;除了 GitHub 和 AI 模型，Vibe Coding 工作流中还有很多环节会受到网络环境影响：

- **npm/pip 包安装**：部分包的下载源在海外，速度受限
- **Docker 镜像拉取**：Docker Hub 访问不稳定
- **技术文档查阅**：很多官方文档托管在海外服务器
- **AI 服务 API**：OpenAI、Anthropic、Google 等 AI 服务的 API 端点

&emsp;&emsp;一个稳定的 VPN 可以一劳永逸地解决这些问题，让你专注于编码本身，而不是与网络环境做斗争。

## 2. 网络环境检测：验证你的 VPN 是否正常工作

&emsp;&emsp;这一章面向**已经有 VPN** 的同学。如果你还没有 VPN，可以跳到下一章查看推荐方案。

&emsp;&emsp;很多同学可能已经有了 VPN，但不确定它是否能满足 Vibe Coding 的需求。接下来我们通过几个简单的测试，验证你的网络环境是否「达标」。<font color=red>如果所有测试都通过，你就可以直接进入 Cursor 和 Git 的安装配置了。</font>

### 2.1 测试一：IP 地区检测（关键！）

&emsp;&emsp;在进行任何服务连通性测试之前，<font color=red>必须先确认你的 VPN 出口 IP 是否在海外地区</font>。这一步非常关键——如果你的 IP 仍然显示为中国地区，即使能访问某些服务，也可能面临账号被封禁的风险。

- **步骤一：打开 VPN 并连接**

&emsp;&emsp;确保你的 VPN 客户端已经连接成功，状态显示为「已连接」。建议选择**美国、日本、新加坡**等节点。

- **步骤二：访问 IP 检测网站**

&emsp;&emsp;打开浏览器，访问 [https://www.ip138.com](https://www.ip138.com)

<div align=center>
  <img src="https://muyu20241105.oss-cn-beijing.aliyuncs.com/images/202602021320642.png" width=80%>
</div>

&emsp;&emsp;**成功判据**：
- 显示的 IP 归属地为**美国、日本、新加坡**等海外地区
- <font color=red>如果显示「中国」，说明 VPN 未生效或节点有问题，必须切换节点后重新测试</font>

- **步骤三：确认地区后继续**

&emsp;&emsp;只有当 IP 归属地显示为海外地区后，才能继续后面的测试。

> &emsp;**为什么这一步这么重要？** OpenAI、Anthropic 等 AI 服务会检测用户 IP 地区。如果使用中国 IP 访问这些服务，轻则无法使用，重则账号被永久封禁。Cursor Pro 订阅的模型调用同样受此影响。

### 2.2 测试二：GitHub 连通性测试

&emsp;&emsp;GitHub 是 Vibe Coding 工作流的核心枢纽，确保它能正常访问。

- **步骤一：浏览器访问 GitHub**

&emsp;&emsp;打开浏览器，访问 [https://github.com](https://github.com)，观察页面加载速度。

&emsp;&emsp;**成功判据**：
- 页面在 3 秒内完全加载
- 图片、头像等资源正常显示
- 可以正常浏览仓库内容

<div align=center>
  <img src="https://muyu20241105.oss-cn-beijing.aliyuncs.com/images/202602021321946.png" width=80%>
</div>


- **步骤二：命令行 Ping 测试**

&emsp;&emsp;打开终端（Windows 用 PowerShell，macOS/Linux 用 Terminal），执行：

```bash
    ping github.com
```

&emsp;&emsp;**成功判据**：
- 有响应（不是 "Request timed out"）
- 延迟在 300ms 以内为佳

> &emsp;注意：部分 VPN 可能会屏蔽 ICMP 协议导致 ping 不通，但实际访问正常。如果 ping 不通但浏览器能正常访问，以浏览器为准。

&emsp;&emsp;完成了网络检测后，如果你发现现有的 VPN 不够稳定，或者你还没有 VPN，下一章我们将推荐一个专为开发者设计的解决方案。
