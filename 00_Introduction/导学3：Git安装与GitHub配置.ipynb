# <center>Git 安装部署与配置指南</center>

&emsp;&emsp;在完成 `Cursor` 的安装与订阅之后，我们的 AI 编程环境其实只完成了一半：IDE 再强，也必须依赖一个稳定的版本控制体系，才能进入真正可协作、可追踪、可交付的工程状态。而在企业研发流程中，`Git` 正是这条主链路的核心基础设施。没有 Git，代码只是散落的文件；有了 Git，代码才能成为可追溯、可协作、可交付的工程资产。

&emsp;&emsp;我们在前面的 Cursor 教学里已经看到，Cursor 的核心优势在于它可以通过 `Agent Mode` 与 `Maximum Context Window` 参与项目级开发。但这里有一个前提：<font color=red>AI 只能在一个规范的工程仓库里发挥价值，而 Git 就是这个工程体系的"骨架"。</font>如果没有正确安装 Git、没有初始化身份、没有配置认证，Cursor 的 Agent 很快就会卡在"无法提交代码""无法拉取仓库""权限认证失败"这些最现实的问题上。这也是为什么我们把 Git 放在 Cursor 之后、作为第二必修课来讲——它是 AI 代码进入真实工程体系的唯一入口。

&emsp;&emsp;因此这一章我们会按完整教学路径展开：首先理解 Git 在 Cursor 工作流中的位置与价值，建立"为什么必须学"的认知基础；然后分别完成 Windows、macOS 与 Linux 的安装部署，确保 Git 在终端与 Cursor 内部都能正常调用；最后统一完成第一次配置（用户名、邮箱、默认分支、Credential Manager），并通过 Cursor 内部的 Git 测试流程，确保你拥有一个真正可用的版本控制环境。

## 1. Git 在 Cursor 编程工作流中的位置

&emsp;&emsp;在进入安装之前，我们必须先回答一个课程级问题：为什么 Git 是 Cursor 环境落地后的第一必修？很多用户会觉得"Git 不就是个版本管理工具吗，以后用到再学"——但如果你真的这样想，后续在 Cursor 里做项目时会频繁踩坑。这一章的作用，就是建立 Git 与 AI IDE 的关系认知，让你理解：<font color=red>Git 不是可选项，而是 Cursor 工程化落地的前置依赖。</font>

&emsp;&emsp;本章将从两个维度展开：首先澄清 Git 的真实定位——它不是"代码备份"，而是工程协作的底层协议；然后说明 Cursor 的高级能力（多文件修改、Agent 自动提交）为什么必须建立在 Git 可用的前提之上。

### 1.1 Git 不是"代码备份"，而是工程协作的底层协议

&emsp;&emsp;很多初学者会把 Git 理解为"版本备份工具"——每次改完代码存一个版本，方便以后回退。这个理解没错，但严重低估了 Git 在企业研发中的真实角色。在真实的工程环境里，Git 承担的是以下四重职责：

- **团队协作协议**：多人同时开发同一个项目时，Git 是唯一能保证代码不冲突、变更可合并的基础设施
- **交付审计链路**：每一次提交（Commit）都带有作者、时间、变更内容，这是企业合规审计的必要数据
- **变更追踪系统**：任何一行代码的修改历史都可以追溯，出了 Bug 可以精准定位是谁、什么时候、为什么改的
- **自动化 CI/CD 的触发源**：现代 DevOps 流程中，代码推送到 Git 仓库会自动触发构建、测试、部署

&emsp;&emsp;Cursor 的 Agent 可以自动生成代码，但这些代码最终必须进入工程体系才能产生价值。而工程体系的入口就是 Git 的四大核心操作：

&emsp;&emsp;<font color=red>所以 Git 是 AI 代码进入真实工程体系的唯一入口。</font>没有 Git，Cursor Agent 生成的代码只能停留在本地文件夹；有了 Git，这些代码才能进入 Commit → Branch → Pull Request → Release Pipeline 的完整工程链路。

### 1.2 Cursor 的高级能力

&emsp;&emsp;Cursor 能做多文件修改、自动重构、跨模块补全，这些能力让它远超传统编辑器。但我们必须认识到，这一切都依赖于一个前提条件：**Git 环境完全可用**。具体来说，Cursor 的以下能力都需要 Git 支撑：

- **仓库结构可读取**：Cursor 的 `Maximum Context Window` 需要读取整个项目结构，而规范的项目结构通常就是一个 Git 仓库
- **变更可提交**：Agent 生成的代码修改需要通过 `git commit` 固化，否则只是临时文件
- **远程可推送**：团队协作、代码审查、CI/CD 触发都需要 `git push` 到远程仓库

&emsp;&emsp;如果 Git 没装好、身份没配置、认证不通过，Cursor 的 Agent 输出就只能停留在"本地草稿"状态，无法进入真正的工程闭环。

## 2. Git 全平台安装部署

&emsp;&emsp;上一章我们理解了 Git 在 Cursor 工作流中的核心位置，知道了它为什么是必修基础设施。那么这一章我们就进入最关键的实战环节：在不同操作系统上完成 Git 的官方安装，并确保它在终端与 Cursor 内部都能正常使用。

&emsp;&emsp;这一章的目标是让大家能够明确三个问题：官方下载入口在哪里？不同操作系统的安装有哪些差异和坑？如何验证安装成功？我们会按 Windows → macOS → Linux 的顺序展开，每个平台都给出完整的步骤和成功判据。

### 2.1 Windows 安装 Git（官方推荐：Git for Windows）

&emsp;&emsp;Windows 下 Git 的官方发行版是 `Git for Windows`，它不仅提供 Git 本体，还附带 `Git Bash`（一个类 Unix 的终端环境）、GUI 工具以及 `Credential Manager`（凭证管理器），是最完整的企业级方案。Git 官方网站：[https://git-scm.com/](https://git-scm.com/)

&emsp;&emsp;<font color=red>注意：不要从第三方网站下载 Git，务必从官网获取最新版本。</font>

- **步骤一：下载安装包**

&emsp;&emsp;打开 Git 官方下载页面：[https://git-scm.com/download/win](https://git-scm.com/download/win)，页面会自动检测你的系统并推荐合适的版本，点击下载最新的 `.exe` 安装器。截至本教程编写时，最新版本为 Git 2.52.0。

### 2.2 macOS 安装 Git（官方推荐：Xcode CLT 或 Homebrew）

&emsp;&emsp;macOS 下 Git 有两条官方推荐路径：一是 Apple 自带的 **Xcode Command Line Tools**（最简单），二是通过 **Homebrew** 安装最新版（推荐企业开发）。Git 官方下载页面：[https://git-scm.com/download/mac](https://git-scm.com/download/mac) 也明确推荐这两种方式。

&emsp;&emsp;对于大多数用户，Xcode CLT 足够使用；但如果你需要最新版 Git 或企业级功能，建议使用 Homebrew 安装。

- **步骤一：安装 Xcode Command Line Tools**

&emsp;&emsp;打开 Terminal（终端），执行以下命令：

```bash
    # 安装 macOS 开发命令行工具（包含 Git）
    xcode-select --install
```

&emsp;&emsp;执行后会弹出安装对话框，点击"安装"即可。整个过程大约需要几分钟，取决于网络速度。安装完成后，Git 就已经可用了。

&emsp;&emsp;**成功判据**：安装完成后，再次运行 `xcode-select --install` 会提示"command line tools are already installed"。

- **步骤二：验证 Git 是否可用**

&emsp;&emsp;在 Terminal 中执行：

```bash
    # 验证 Git 版本
    git --version
```

&emsp;&emsp;**成功判据**：输出类似 `git version 2.39.3 (Apple Git-145)` 的版本信息。注意 Apple 内置的 Git 版本通常比最新版略旧，但对于日常开发完全够用。

- **步骤三（推荐）：使用 Homebrew 安装最新版 Git**

&emsp;&emsp;如果你需要最新版 Git（企业开发推荐），可以通过 Homebrew 安装。首先确保已安装 Homebrew，然后执行：

```bash
    # 使用 Homebrew 安装最新版 Git
    brew install git

    # 验证安装的是 Homebrew 版本而非 Apple 版本
    which git
    git --version
```

&emsp;&emsp;安装完成后，`which git` 应该输出 `/opt/homebrew/bin/git`（Apple Silicon）或 `/usr/local/bin/git`（Intel），而不是 `/usr/bin/git`（Apple 内置版本）。

### 2.3 Linux 安装 Git（Ubuntu/Debian/Fedora）

&emsp;&emsp;Linux 是服务器与 DevOps 环境的主战场，Git 几乎永远通过系统包管理器安装。Git 官方下载页面：[https://git-scm.com/download/linux](https://git-scm.com/download/linux) 列出了各发行版的安装命令。不同发行版使用不同的包管理器，但安装都非常简单。

- **步骤一：Ubuntu/Debian 系统安装**

&emsp;&emsp;Ubuntu 和 Debian 使用 `apt` 包管理器，执行以下命令：

```bash
    # 更新 apt 源索引
    sudo apt-get update

    # 安装 Git 完整版（包含所有依赖）
    sudo apt-get install git-all -y
```

&emsp;&emsp;`git-all` 包含了 Git 的所有组件，包括 GUI 工具和文档。如果只需要核心功能，可以改用 `sudo apt-get install git`。

- **步骤二：Fedora/CentOS/RHEL 系统安装**

&emsp;&emsp;这些发行版使用 `dnf`（或旧版的 `yum`）包管理器：

```bash
    # Fedora/CentOS 安装 Git
    sudo dnf install git-all -y
```

- **步骤三：验证安装成功**

&emsp;&emsp;无论使用哪个发行版，安装完成后都用以下命令验证：

```bash
    # 验证 Git 版本
    git --version
```

&emsp;&emsp;**成功判据**：输出类似 `git version 2.43.0` 的版本信息。Linux 仓库中的 Git 版本通常比较新，企业生产环境完全可用。

> &emsp;**小结**：这一章我们完成了 Git 在 Windows/macOS/Linux 三大平台的安装部署。但"装好了"不等于"可用"——Git 还需要一个远程代码托管平台来实现团队协作与代码备份。下一章，我们就来注册 GitHub 账户，打通本地与远程的连接通道。

## 3. GitHub 账户注册与 Token 配置

&emsp;&emsp;Git 本身是一个本地版本控制工具，但真正的工程协作需要一个**远程代码托管平台**来存储代码、管理权限、触发 CI/CD。在全球范围内，`GitHub` 是最主流的选择——它不仅是代码托管平台，更是全球最大的开发者社区和开源生态中心。

&emsp;&emsp;对于 Cursor 用户来说，GitHub 账户还有一个关键作用：<font color=red>Cursor 的 Agent 在执行 `git push` 时需要通过 GitHub 认证，而认证的核心凭证就是 Personal Access Token（PAT）。</font>没有 GitHub 账户和 Token，你的代码就只能停留在本地，无法进入远程协作流程。

&emsp;&emsp;这一章我们会完成两件事：首先注册 GitHub 账户并完成邮箱验证；然后创建 Personal Access Token，为后续的 Git 认证配置做好准备。

### 3.1 注册 GitHub 账户

&emsp;&emsp;GitHub 的注册流程非常简单，只需要一个有效的邮箱地址。<font color=red>建议使用常用邮箱（如 Gmail、Outlook、QQ邮箱等），因为这个邮箱会成为你在 Git 提交记录中的身份标识。</font>

- **步骤一：访问 GitHub 官网**

&emsp;&emsp;打开浏览器，访问 GitHub 官网：[https://github.com/](https://github.com/)，点击右上角的 **Sign up** 按钮进入注册页面。

<div align=center>
  <img src="https://muyu20241105.oss-cn-beijing.aliyuncs.com/images/202602021233712.png" width=80%>
</div>

- **步骤二：填写注册信息**

&emsp;&emsp;GitHub 会引导你依次填写以下信息：

1. **Email**：输入你的常用邮箱地址
2. **Password**：设置一个强密码（至少 8 位，包含数字和小写字母）
3. **Username**：选择一个用户名，这将成为你的 GitHub 主页地址（如 `github.com/your-username`）
4. **Email preferences**：是否接收产品更新邮件（可选）
5. **Verify**：完成人机验证（通常是图片验证或拼图验证）

&emsp;&emsp;填写完成后，点击 **Create account** 按钮。

<div align=center>
  <img src="https://muyu20241105.oss-cn-beijing.aliyuncs.com/images/202602021216458.png" width=80%>
</div>

- **步骤三：验证邮箱**

&emsp;&emsp;注册完成后，GitHub 会向你填写的邮箱发送一封验证邮件。打开邮箱，找到来自 GitHub 的邮件，点击邮件中的 **Verify email address** 按钮完成验证。

&emsp;&emsp;<font color=red>注意：如果没有收到邮件，请检查垃圾邮件文件夹。邮箱验证是必须的，否则部分功能会受限。</font>

- **步骤四：完成初始化设置（可跳过）**

&emsp;&emsp;邮箱验证通过后，GitHub 会引导你完成一些初始化设置：

1. **个性化问卷**：询问你的团队规模、角色、使用目的等（可以选择 Skip 跳过）
2. **选择计划**：GitHub 提供 Free 和 Pro 两种计划，个人开发者选择 **Free** 即可，功能完全够用
3. **创建第一个仓库**：可以暂时跳过，我们后续会在 Cursor 中创建

&emsp;&emsp;完成这些步骤后，你就拥有了一个完整的 GitHub 账户！

### 3.2 创建 Personal Access Token（PAT）

&emsp;&emsp;GitHub 从 2021 年开始，已经**禁止使用密码进行 Git 命令行认证**，必须使用 Personal Access Token（个人访问令牌）。这个 Token 就是你通过命令行操作 GitHub 仓库的"通行证"。

&emsp;&emsp;<font color=red>重要：Token 只会在创建时显示一次，之后无法再次查看。请务必在创建后立即保存到安全的地方！</font>


- **步骤一：进入 Settings**

&emsp;&emsp;登录 GitHub 后，点击右上角的头像，在下拉菜单中选择 **Settings**。

<div align=center>
  <img src="https://muyu20241105.oss-cn-beijing.aliyuncs.com/images/202602021216460.png" width=60%>
</div>

- **步骤二：进入 Developer settings**

&emsp;&emsp;在 Settings 页面的左侧边栏中，滚动到最底部，点击 **Developer settings**。

<div align=center>
  <img src="https://muyu20241105.oss-cn-beijing.aliyuncs.com/images/202602021216461.png" width=80%>
</div>

- **步骤三：创建新 Token**

&emsp;&emsp;在 Developer settings 页面：

1. 点击左侧的 **Personal access tokens** → **Tokens (classic)**
2. 点击 **Generate new token** → **Generate new token (classic)**
3. 填写 Token 信息：
   - **Note**：给 Token 起个名字，如 `Cursor-Git-Token`
   - **Expiration**：选择有效期（建议选择 90 days 或 No expiration）
   - **Select scopes**：勾选权限范围

## 4. Git 初始化配置与 Cursor 集成验证

&emsp;&emsp;上一章我们完成了 GitHub 账户注册和 Token 创建，现在手握"通行证"，但 Git 本身还没有完成**身份初始化**——如果你现在直接尝试 `git commit`，Git 会拒绝并提示"Please tell me who you are"。

&emsp;&emsp;这一章我们会完成三件事：首先配置 Git 的全局身份（用户名和邮箱）；然后设置 Credential Manager 并绑定上一章创建的 Token，解决反复输入密码的问题；最后在 Cursor 内部完成一次完整的 Git 操作测试，确保开发闭环真正可用。

&emsp;&emsp;我们要验证 Cursor Agent 与 Git 的**完整协作闪环**：在 GitHub 创建空仓库 → 在 Cursor Agent 中一句话完成代码创建与推送。这个流程不仅验证了 Git 配置是否正确，更展示了 AI 编程的真正威力——<font color=red>你只需要告诉 Agent 做什么，它会自动完成 git init、commit、push 全流程。</font>
